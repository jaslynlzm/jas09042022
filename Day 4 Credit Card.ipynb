```python
import pandas as pd
```

```python
df = pd.read_csv("Credit Card Default II balance.csv")
```

```python
df = df.loc[:, ["default", "income", "age", "loan"]]
```

```python
df = df.dropna()
```

```python
X = df.loc[:, ["income", "age", "loan"]]
```

```python
Y = df.loc[:, ["default"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state = 204)
```

```python
from sklearn import tree
```

```python
model1 = tree.DecisionTreeClassifier(min_samples_split=250, random_state=204)
```

```python
model1.fit(X_train,Y_train)
```

```python
pred = model1.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_train, pred)
```

```python
cm
```

```python
accuracy = (cm[0,0] + cm[1,1])/(sum(sum(cm)))
```

```python
accuracy
```

```python
pred = model1.predict(X_test)
cm = confusion_matrix(Y_test, pred)
accuracy = (cm[0,0] + cm[1,1])/(sum(sum(cm)))
```
